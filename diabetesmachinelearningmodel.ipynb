```python
# importing all necessary libraries. scikit is amachine learning library that supplies data sets and regression algorithms
# we will be utilizing supervised machine learning
import matplotlib.pyplot as plt
import numpy as np
# model_selection is used to to split data into training and test sets
from sklearn import datasets, linear_model, model_selection
```

```python
# loading the dataset. The command "return_X_y = True" will signal that X is a data matrix while y is a regression target.
X, y = datasets.load_diabetes(return_X_y= True )

# Will display 422 items shaped in arrays of 10 elements
print(X.shape)

# What does the first row look like?
print(X[0])
```

```python
# Selecting the 2nd column over the data matrix (bmi)
X = X[:, np.newaxis, 2]


```

```python
# Splitting the data into test and training sets. Test size = 0.33 means that 33 percent of data will be in test part and the rest will be in train part.
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# Finding a linear regression between the data. Train the model with your training sets
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predicting the y values (the line)
y_pred = model.predict(X_test)
```

```python
# Plotting!
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
