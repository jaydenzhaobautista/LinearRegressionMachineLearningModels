```python
# importing all necessary libraries. scikit is amachine learning library that supplies data sets and regression algorithms
# we will be utilizing supervised machine learning
import matplotlib.pyplot as plt
import numpy as np
# model_selection is used to to split data into training and test sets
from sklearn import datasets, linear_model, model_selection

# loading the specific dataset. In this model we will be find a line of best fit for the Linnerrud dataset. The x variable will be the waist, and the y variable will be the number of situps
from sklearn.datasets import load_linnerud
```

```python
# grouping the dataset into X and y values. This is a DataFrame of 20 rows by 6 columns. First 3 columns represent the feature variables (weight, waist, and pulse), while the last three columns represent the target variables (chins, situps, and jumps)
X, y = load_linnerud(return_X_y=True)


# Both are tuples of 20 rows by 3 columns
print(X.shape)
print(y.shape)
```

```python
# Selecting the 2nd columns of each dataset, which is waistline vs number of situps
X = X[:, np.newaxis, 1]
y = y[:, np.newaxis, 1]
```

```python
# Splitting the data into test and training sets. Test size = 0.33 means that 33 percent of data will be in test part and the rest will be in train part.
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# Finding a linear regression between the data. Train the model with your training sets
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predicting the y values (the line)
y_pred = model.predict(X_test)
```

```python
# Plotting!
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Waistline')
plt.ylabel('Number of Situps')
plt.title('A Graph Plot Showing the Relationship Between the Number of Situps and Waistline')
plt.show()
```
